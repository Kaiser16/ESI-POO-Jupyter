```c++
# include <iostream>

class Persona
{
public:
    Persona(std::string nombre,unsigned edad): nombre_(nombre),edad_(edad)
    {
        vivo = true;
    };
    const std::string nombre() const {return nombre_;};
    const unsigned edad() const {return edad_;};
    void crecer()
    {
        if(!vivo)
        {
            std::cout<<nombre_<<" ya estaba F"<<std::endl;
        }
        else
        {
            ++edad_;
            if(edad_ >= 100)
            {
                vivo = false;
                std::cout<<"F "<<nombre_<<std::endl;
            }
            else
            {
                std::cout<<nombre_<<" tiene ";
                std::cout<<edad_<<" años."<<std::endl;
            }
        }
    };
private:
    std::string nombre_;
    unsigned edad_;
    bool vivo;
}
```

```c++
Persona p1("Manolo",1)
```

```c++
for(int i = 0; i < 100; ++i)
    p1.crecer()
```
